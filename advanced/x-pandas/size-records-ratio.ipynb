```python
# Importing Pandas to create DataFrame
import pandas as pd
  
df = pd.read_csv("connecctions_df.csv")

print(df)
```

```python
name_hash_size_df = pd.read_csv("name_hash_filesize.csv")
name_hash_size_df
```

```python
name_hash_events = df.drop(columns=["Unnamed: 0","file_creation_ts","file_type","direction","date"])
name_hash_events
```

```python
df_nhecsb = pd.merge(name_hash_events,name_hash_size_df)
```

```python
df_nhecsb["sender"] = df_nhecsb["file_name"].str[2:7]
df_nhecsb["recipient"] = df_nhecsb["file_name"].str[7:12]
df_nhecsb["seqno"]= df_nhecsb["file_name"].str[12:]
df_nhecsb["prefix"]= df_nhecsb["file_name"].str[:2]
nrt_df = df_nhecsb.loc[df_nhecsb["prefix"]=="AB"]
```

```python
nrt_df[["sender","file_size_bytes","events_count"]].to_csv("data_to_learn.csv")
```

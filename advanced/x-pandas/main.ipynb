## Filtering Data

```python
new_df = data.loc[ (data[" file_type"]=="NRT") & (data[" file_count"] > 100) ]
new_df.reset_index(drop=True,inplace=True)
new_df
```

## Filter on column contains str

```python
newDF.loc[newDF["connection"].str.contains("GEOGC")]
```

## Filter with reguler expressions

```python
import re
newDF.loc[newDF["connection"].str.contains("JPN[A-Z0-9]{2}",flags=re.I,regex=True)]
```

## Conditional changes to data

```python
data.loc[data["connection"].str.startswith("ABqqqqq99999"), "connection"] = "ABqqqqq99999!!"
data.loc[data["connection"].str.startswith("ABqqqqq99999"), " file_type"] = "VVV"
data[data["connection"].str.startswith("ABqqqqq99999")]
data.drop(columns=["file_type"])
```

```python
data.loc[ data["EventsPerFile"] > 15000, ["Flag","Case"]]="CashCow"
data[data["Flag"]=="CashCow"].sort_values("EventsPerFile",ascending=False)
```

## Aggregate Statistics (Group by)

```python
data.groupby(["connection"," file_type"])[[" events_sum","EventsPerFile"]].count().sort_values(" events_sum",ascending=False)
```

```python
data.groupby(["connection"," file_type"])[[" events_sum","EventsPerFile"]].sum().sort_values(" events_sum",ascending=False)
```

```python
data.groupby(["connection"," file_type"])[[" events_sum","EventsPerFile"]].mean().sort_values(" events_sum",ascending=False)
```

```python
import datetime
from typing import Any


def is_before(a: Any, b: Any) -> bool:
    def conv(x: Any) -> datetime.datetime:
        if isinstance(x, datetime.datetime):
            return x if x.tzinfo else x.replace(tzinfo=datetime.timezone.utc)
        if isinstance(x, datetime.date):
            return datetime.datetime(
                x.year, x.month, x.day, tzinfo=datetime.timezone.utc
            )
        if isinstance(x, (int, float)) and not isinstance(x, bool):
            return datetime.datetime.fromtimestamp(x, tz=datetime.timezone.utc)
        if isinstance(x, str):
            s = x.strip().replace("Z", "+00:00")
            try:
                d = datetime.datetime.fromisoformat(s)
            except ValueError:
                d = datetime.datetime.strptime(s, "%Y-%m-%d %H:%M:%S")
            return d if d.tzinfo else d.replace(tzinfo=datetime.timezone.utc)
        raise TypeError(type(x))

    return conv(a) < conv(b)
```

```python
is_before("2025-01-01", "2024-01-01")
```

```python
is_before("2025-01-01 10:00:00", "2024-01-01")
```

```python
is_before(12700000, "2024-01-01")
```

```python
is_before("2025-01-01 10:00:00", datetime.datetime(2026, 1, 1))
```

```python
is_before("2025-01-01 10:00:00", datetime.datetime(2026, 1, 1, tzinfo=datetime.timezone.utc))
```

```python
# Importing Pandas to create DataFrame
import pandas as pd
  
df = pd.read_csv("connecctions_df.csv")

print(df)
df.loc[df["file_name"]=="file_name_goes_here"]
```

```python
search_for = [
    "file_name1",
    "file_name2",
    "file_name3",
    "file_name4",
    "file_name5",
    "file_name6",
]
missing =set(search_for).difference(set(df.loc[df["file_name"].isin(search_for)].file_name.to_list()))
list(missing)
```

```python
df["file_type"]=df.apply(lambda row:"value1" if row["file_name"][:2]=="AB" else "value2",axis=1)
df["direction"]=df.apply(lambda row:"OUT" if row["file_name"][2:7] == "97865" else "IN",axis=1)
df.columns
```

```python
connection_pivot = pd.pivot_table(df,
                                 values = 'events_count',
                                 index='date',
                                 columns = ['file_type','direction'],
                                 aggfunc=["sum"],
                                 fill_value=0.0
                                 ).reset_index()
print(connection_pivot)
```
